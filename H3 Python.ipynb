```python
lostsuperreal = lostsuper[lostsuper.Population.notnull()]
```

```python
import sklearn.cluster
import sklearn.metrics
%matplotlib inline
import matplotlib
import numpy
import sklearn.neighbors
```

```python
import sklearn.preprocessing
label_enc = sklearn.preprocessing.LabelEncoder()
label_enc.fit_transform(lostsuperreal3.State)
```

```python
lostsuperreal3.State.unique()
```

```python
lostsuperreal3['State_as_a_number'] = label_enc.fit_transform(lostsuperreal3.State)
lostsuperreal3.State_as_a_number
```

```python
df = lostsuperreal3[['LS_Value_Per_Person', 'Tax_Median_Income', 'State_as_a_number']]
df
```

```python
clusterer = sklearn.cluster.KMeans(n_clusters=3)
```

```python
cluster_suggestions = clusterer.fit_predict(df)
cluster_suggestions
```

```python
sklearn.metrics.silhouette_score(df, cluster_suggestions)
```

```python
cluster5 = sklearn.cluster.KMeans(n_clusters=5)
```

```python
cluster_suggestions5 = clusterer.fit_predict(df)
cluster_suggestions5
```

```python
sklearn.metrics.silhouette_score(df, cluster_suggestions5)
```

```python
scores = []
for i in range(2, 40):
    clustereri = sklearn.cluster.KMeans(n_clusters=i)
    cluster_suggestionsi = clustereri.fit_predict(df)
    scores.append(sklearn.metrics.silhouette_score(df, cluster_suggestionsi))
scores
```

```python
matplotlib.pyplot.plot(range(2,40), scores)
```

```python
cluster4 = sklearn.cluster.KMeans(n_clusters=4)
cluster_suggestions4 = cluster4.fit_predict(df)
cluster_suggestions4
```

```python
cluster4.cluster_centers_
```

```python
X = df['Tax_Median_Income']
y = df['LS_Value_Per_Person']
```

```python
colouring = numpy.array(['purple', 'pink', 'orange', 'green'])
```

```python
colour_by_cluster = colouring[cluster_suggestions4]
colour_by_cluster
```

```python
matplotlib.pyplot.scatter(X,y, c=colour_by_cluster)
```

KMeans cant really cluster this. The data is more linear and not suitable to clustering

```python
X = df['State_as_a_number']
y = df['LS_Value_Per_Person']
```

```python
colouring = numpy.array(['purple', 'pink', 'orange', 'green'])
colour_by_cluster = colouring[cluster_suggestions4]
colour_by_cluster
```
